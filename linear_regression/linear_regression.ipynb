```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# load data set
df = pd.read_csv('salary_data.csv')
df.head()
```

```python
# renaming columns for consistency 
df = df.rename(columns={"Years of Experience": 'YearsExperience'})
```

```python
# feature selection
x = df[['YearsExperience']]
y = df['Salary']
```

```python
# model selection and fitting
model = LinearRegression()
model.fit(x.values, y.values)
```

```python
df['PredictedSalary'] = model.predict(x.values)
```

```python
print(f" model slop: {round(model.coef_[0], 2)}")
print(f" model intercept: {round(model.intercept_,2)}")
```

```python
# plotting the results
plt.scatter(df['YearsExperience'], df['Salary'], color='blue', label='Actual Salary')
plt.plot(df['YearsExperience'], df['PredictedSalary'], color='red', label='Predicted Salary')
plt.xlabel('Years of Experience')  
plt.title('Salary vs Years of Experience')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show() 
```
